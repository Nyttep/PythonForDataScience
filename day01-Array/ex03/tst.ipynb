```python
from load_image import ft_load
import numpy as np
from PIL import Image

img_arr = ft_load("animal.jpeg")
```

```python
img = Image.fromarray(img_arr).crop((0, 0, 400, 400))
print(img.size)
img.show()
```

```python
grayscale = img.convert(mode="L")
grayscale.show()
```

```python
import matplotlib.pyplot as plt

plt.imshow(grayscale)
plt.show()
```
